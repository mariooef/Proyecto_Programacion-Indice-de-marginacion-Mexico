> Se realiza una gráfica que permite ver el porcentaje de municipios por estado con índices de marginación "muy bajo", "bajo", "medio","alto" y "muy alto".
>
> La grafica utilizada para representar la información solicitada es la de barras apiladas, siendo que es la que más se ajusta a representar los datos y darnos un panorama real visualmente hablando de como está distribuido el porcentaje de marginación entre los distintos estados.



```python
# Se define el orden de los grados de marginación, esto para poder representarlos del nivel más bajo al más alto en la gráfica.
order = ["Muy bajo", "Bajo", "Medio", "Alto", "Muy alto"]
df_marginacion['GM_2020'] = pd.Categorical(df_marginacion['GM_2020'], categories=order, ordered=True)

# Se agrupa por estado y grado de marginación, para así contar el número de municipios en cada grupo y asi Calcular el porcentaje de municipios por grado de marginación en cada estado.
grouped = df_marginacion.groupby(['NOM_ENT', 'GM_2020']).size().unstack()
percentages = grouped.div(grouped.sum(axis=1), axis=0) * 100

# Se crea un diccionario de colores para cada grado de marginación, esto para representar los colores más adecuados visualmente al nivel de marginación.
colormap = {
    'Muy bajo': '#65D5FF',
    'Bajo': '#67FF65',
    'Medio': '#FFFA65',
    'Alto': '#FFAB65',
    'Muy alto': '#FF6565'
}

# Se crea la gráfica de barras apiladas
ax = percentages.plot(kind='bar', stacked=True, figsize=(10, 6), color = colormap)

ax.set_title('Porcentaje de municipios por estado y grado de marginación')
ax.set_xlabel('Estado')
ax.set_ylabel('Porcentaje')

ax.legend(title='Grado de Marginación')

# Se Agrega el símbolo de porcentaje a las etiquetas del eje y
ax.set_yticklabels(['{:.0f}%'.format(x) for x in ax.get_yticks()])

# Se guarda la gráfica en un archivo PNG
pl.savefig('grafica_marginacion.png')

pl.show()
```

> Se realiza una gráfica que muestra el porcentaje de la población, respecto a la población total de cada estado, con índices de marginación "muy bajo", "bajo", "medio", "alto" y "muy alto".
>
> Aquí se decidio utilizar el mismo tipo de grafica de tipo barra apilada, esto para poder hacer una comparación de los datos con la anterior grafica e igual se ajusta a lo que se quiere representar, mostrando así como la marginación se distribuye porcentualmente entre el porcentaje de población respecto a la total.

```python
# Se agrupa por estado y grado de marginación, logrando con ello la suma de la población total en cada grupo
grouped_pop = df_marginacion.groupby(['NOM_ENT', 'GM_2020'])['POB_TOT'].sum().unstack()

# Se calcula el porcentaje de población por grado de marginación en cada estado
percentages_pop = grouped_pop.div(grouped_pop.sum(axis=1), axis=0) * 100

ax = percentages_pop.plot(kind='bar', stacked=True, figsize=(10, 6), color=[colormap.get(c) for c in percentages_pop.columns])

# Configuraciones de la grafica
ax.set_title('Porcentaje de población por estado y grado de marginación')
ax.set_xlabel('Estado')
ax.set_ylabel('Porcentaje de Población')

ax.legend(title='Grado de Marginación')

ax.set_yticklabels(['{:.0f}%'.format(x) for x in ax.get_yticks()])

pl.savefig('grafica_poblacion.jpg')

pl.show()
```

**Hallazgos estadísticos**

* De manera general se puede decir que no hay coincidencias, entre los datos presentados en las dos gráficas, salvo por uno o dos estados que en sus datos conservan una relación entre las dos graficas. Se puede observar que el grado nivel muy bajo aumento en la segunda gráfica, es decir el grado de marginación a niveles muy bajo se encuentra en el mayor porcentaje de población a nivel estado con respecto al total de la población y en la primera grafica observamos que el grado de marginación tiene niveles más altos con respecto al porcentaje de municipios de cada estado.

* Existen estados que se podría decir cuentan con índices muy bajos de marginalidad entre las dos graficas en cual no les afecta si es por municipio o por porcentaje poblacional.

> Se Desarrollo un nuevo DataFrame con indicadores interesantes por estado que se pueden obtener de los datos a nivel municipal y que pudieran ser de importancia a la hora de definir políticas públicas este nuevo DataFrame se guarda en formato parquet.

```python
# Crear un nuevo DataFrame con las columnas existentes
df_nuevos_indicadores = df_marginacion[['CVE_ENT', 'NOM_ENT', 'CVE_MUN', 'NOM_MUN', 'POB_TOT']].copy()

# Agregar nuevos indicadores al nuevo DataFrame
df_nuevos_indicadores['VIPE.2KM'] = pd.Series(dtype=float)  # Porcentaje de vivienda que cuenta con a alguna institución publica educativa a menos de 2 km de distancia
df_nuevos_indicadores['OVAI'] = pd.Series(dtype=float)  # Porcentaje de ocupantes en viviendas particulares con acceso a internet
df_nuevos_indicadores['COMDIA'] = pd.Series(dtype=float)  # Porcentaje de la población que tiene disponibilidad de no más 2 comidas al día.
df_nuevos_indicadores['EMNA'] = pd.Series(dtype=float)  # Porcentaje de muertes no accidentales con edad menor de 65 años.
df_nuevos_indicadores['OME.2'] = pd.Series(dtype=float)  # Porcentaje de ocupantes menores de edad en viviendas particulares que superan la cantidad de 2

# Guardar el nuevo DataFrame en formato parquet
df_nuevos_indicadores.to_parquet('Nuevos_Indicadores_Dataframe_IMM_2020.parquet')
```

**Justificación de los indicadores**

* **VIPE.2KM** = Este indicador es importante para saber que tan disponible, sin ningún otro factor, alguna escuela o institución publica educativa independientemente del nivel. Por qué elegí 2 kilómetros, me parece una distancia razonable en donde el factor de trasporte no debe influir y tampoco estaría lo demasiado cerca para afectar alguna zona habitacional provocando otros problemas viales o civiles. Este indicador puede ayudar a en caso de que el porcentaje en una localidad sea bajo, puede dar la pauta de construir escuelas más cercanas o bien optar por un sistema de transporte especial para estudiantes y así permitan a cada familia quitar ese factor para la educación, el transporte cuesta y esto puede ser causa de deserción escolar.

* **OVAI** = Este indicador, aunque pueda parecer incongruente con los índices de marginalidad, puesto que el acceso a internet se puede seguir interpretando como no básico, si se me ha hecho interesante preguntarme realmente la población con índices de marginalidad altos o medios que tanto acceso a internet tienen (vía celular, vía telefónica, etc) y si a pesar de índices altos el internet al final responde la pregunta que se ha convertido en parte importante para la población. Este indicador también puede dar la pauta para poder determinar de qué manera se educa o bien se informa o se mantiene en contacto con la población marginada, si existe un medio de comunicación es más fácil llegar a ellos y ayudarlos.

* **COMDIA** = Este indicador podría ser de mucha ayuda para generar programas adecuados para otorgar acceso a una mejor alimentación, ya sea para comedores comunitarios o bien para acceso a vales de despesa por ley (que puedan ser cambiados por ciertas tiendas y con reglas para su uso formal). Todo ser humano o bien toda persona que forme parte de una población debe tener acceso al menos a 3 comidas diarias, este indicador lógicamente no estaría contando a personas que por decisión deciden hacer una o dos o ninguna comida, es porcentaje de cuantas comidas al día se tiene acceso.

* **EMNA** = Este indicador indica el porcentaje de muertes en personas menores de 65 años que no tenga que estar relacionado con accidentes. Muertes naturales, enfermedad, delincuencia. es importante conocer este indicador para saber que tanto es la esperanza de vida y si la marginalidad influye en ella. esto se podría tomar para mejorar el sistema de salud y los servicios médicos en general en localidad que así lo requiera en donde la marginalidad y la esperanza de vida estén de la mano.

* **OME.2** = La justificación para este indicador es que es importante conocer que tanto está relacionado el porcentaje de personas menores edad que superan los 2 habitantes por vivienda con el índice de marginalidad y si hay una relación entre más índice de marginalidad. esto ayudara para crear programas más eficientes para el control de natalidad o bien crear leyes que puedan controlar el número de nacimientos permitidos (esto es algo extremista, pero entre más personas que no tengan acceso a recursos suficientes o trabajos formales que le permitan tener una calidad de vida adecuada, el índice de marginalidad seguirá en ascenso).
